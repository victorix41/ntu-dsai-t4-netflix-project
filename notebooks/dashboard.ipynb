```python
mkdir dashboard
cd dashboard
```

### STEP 2: Create environment.yml

```python
name: dashboard
channels:
  - conda-forge
  - defaults
dependencies:
  - python=3.11
  - pandas
  - pip
  - pip:
      - streamlit
      - plotly
```

### STEP 3: In Terminal, create a fresh conda environment (one-time only)

```python
conda env create -f environment.yml
```

### STEP 4: Activate conda streamlit environment in the folder (just do this if environment.yml is done)

```python
conda activate dashboard
```

### STEP 5: Install dependencies
